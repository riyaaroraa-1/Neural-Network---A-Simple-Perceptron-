**Question 1:  What is Deep Learning? Briefly describe how it evolved and how it differs
from traditional machine learning.**

**Answer:**
Deep Learning is a branch of Machine Learning that uses Artificial Neural Networks (ANNs) with multiple hidden layers to automatically learn patterns and features from large amounts of data. It is capable of solving complex problems such as image recognition, speech recognition, natural language processing, and autonomous driving without requiring extensive manual feature engineering.

**Evolution of Deep Learning**

Deep Learning evolved from the concept of artificial neural networks inspired by the human brain. Early neural network models were introduced in the 1950s, but their practical use was limited due to insufficient computing power and lack of large datasets. With advancements in powerful GPUs, the availability of big data, and improved training algorithms such as backpropagation, deep learning gained popularity in the 2010s. Today, it is widely used in applications such as computer vision, natural language processing, healthcare, and recommendation systems.

**Difference between Deep Learning and Traditional Machine Learning**

Traditional Machine Learning relies on manual feature extraction, where domain experts identify important features before training the model. It generally performs well on structured datasets and requires less computational power.

In contrast, Deep Learning automatically learns relevant features directly from raw data using multiple neural network layers. It performs exceptionally well on unstructured data such as images, audio, and text but requires large datasets, high computational resources, and longer training times.


**Question 2: Explain the basic architecture and functioning of a Perceptron. What are its
limitations?**

**Answer:**

A Perceptron is the simplest type of Artificial Neural Network (ANN) and serves as the basic building block of deep learning. It is a supervised learning algorithm used for binary classification problems, where the output belongs to one of two classes.

**Basic Architecture of a Perceptron**

A perceptron consists of the following components:

* **Input Layer:** Receives input features (x₁, x₂, ..., xₙ).
* **Weights:** Each input is assigned a weight (w₁, w₂, ..., wₙ) that represents its importance.
* **Bias:** A bias term is added to shift the decision boundary and improve the model's flexibility.
* **Summation Function:** Computes the weighted sum of the inputs and bias.
* **Activation Function:** Applies a threshold function to produce the final output, which is typically 0 or 1.

**Functioning of a Perceptron**

1. The input features are multiplied by their corresponding weights.
2. The weighted inputs are summed together, and the bias is added.
3. The result is passed through an activation function (such as the step function).
4. If the output exceeds a predefined threshold, the perceptron predicts one class (1); otherwise, it predicts the other class (0).
5. During training, the weights and bias are updated based on the prediction error until the model learns the correct decision boundary.

**Limitations of a Perceptron**

* It can solve only **linearly separable** problems.
* It cannot solve complex problems such as the **XOR** problem.
* It uses a simple activation function, which limits its learning capability.
* It cannot learn complex patterns present in high-dimensional or non-linear data.
* Its performance is limited compared to multi-layer neural networks used in deep learning.


**Question 3: Describe the purpose of activation function in neural networks. Compare
Sigmoid, ReLU, and Tanh functions.**

**Answer:**

An activation function is a mathematical function applied to the output of a neuron in a neural network. Its main purpose is to introduce non-linearity, enabling the network to learn complex patterns and relationships in the data. Without activation functions, a neural network would behave like a simple linear model and would not be able to solve complex tasks such as image recognition, speech recognition, and natural language processing.

**Sigmoid Function:**
The Sigmoid function converts input values into an output ranging from 0 to 1. It is commonly used in the output layer of binary classification problems because its output can be interpreted as a probability. However, it suffers from the vanishing gradient problem, where gradients become very small during training, making learning slow in deep neural networks.

**Tanh (Hyperbolic Tangent) Function:**
The Tanh function produces output values between -1 and 1. Unlike the Sigmoid function, it is zero-centered, which helps the network learn more efficiently. Although it generally performs better than Sigmoid, it also suffers from the vanishing gradient problem, making it less suitable for very deep neural networks.

**ReLU (Rectified Linear Unit) Function:**
The ReLU function outputs the input value if it is positive; otherwise, it returns zero. It is computationally simple, speeds up training, and largely overcomes the vanishing gradient problem. Due to these advantages, ReLU is the most widely used activation function in the hidden layers of deep neural networks. However, it may suffer from the dying ReLU problem, where some neurons stop updating if they consistently receive negative inputs.

**Difference between Sigmoid, Tanh, and ReLU:**

* **Sigmoid** produces output values between 0 and 1 and is mainly used for binary classification, but it suffers from slow training due to the vanishing gradient problem.
* **Tanh** produces output values between -1 and 1, is zero-centered, and usually performs better than Sigmoid, but it also experiences the vanishing gradient problem.
* **ReLU** produces output values from 0 to positive infinity, trains faster, is computationally efficient, and is the preferred activation function for hidden layers in deep neural networks, although it may suffer from the dying ReLU problem.


**Question 4: What is the difference between Loss function and Cost function in neural
networks? Provide examples.**

**Answer:**
In neural networks, **Loss Function** and **Cost Function** are used to measure the error between the predicted output and the actual output. Although the terms are often used interchangeably, they have slightly different meanings.

A **Loss Function** calculates the error for a single training example. It measures how far the model's prediction is from the actual value for one data point. During training, the loss is computed for each individual sample.

A **Cost Function**, on the other hand, calculates the average or total loss over the entire training dataset. It represents the overall performance of the model and is the value that optimization algorithms, such as Gradient Descent, aim to minimize by updating the model's weights.

**Difference between Loss Function and Cost Function**

* Loss Function measures the error for a single training example, whereas the Cost Function measures the average or total error across the entire dataset.
* The Loss Function evaluates the prediction of one data point, while the Cost Function evaluates the overall performance of the model during training.
* During optimization, individual losses are combined to compute the cost, and the model parameters are updated to minimize this cost.

**Examples**

* **Mean Squared Error (MSE):** Commonly used for regression problems. It calculates the average of the squared differences between actual and predicted values. When computed for one sample, it represents a loss; when averaged over all samples, it becomes the cost.

* **Binary Cross-Entropy Loss:** Used for binary classification problems. It measures the difference between the predicted probability and the actual class label.

* **Categorical Cross-Entropy Loss:** Used for multi-class classification problems where the output belongs to one of several classes.



**Question 5: What is the role of optimizers in neural networks? Compare Gradient
Descent, Adam, and RMSprop.**

**Answer:**

An optimizer is an algorithm used to update the weights and biases of a neural network during training. Its main role is to minimize the loss (or cost) function by adjusting the model parameters so that the predictions become more accurate. Optimizers determine how the network learns from errors and improve the model's performance over successive training iterations.

The most commonly used optimizers are **Gradient Descent, RMSprop, and Adam**.

**Gradient Descent:**
Gradient Descent updates the model parameters by moving them in the direction opposite to the gradient of the loss function. It uses a fixed learning rate to minimize the error. Although it is simple and widely used, it can be slow to converge and may get stuck in local minima or oscillate if the learning rate is not chosen properly.

**RMSprop (Root Mean Square Propagation):**
RMSprop improves Gradient Descent by adapting the learning rate for each parameter individually. It uses the moving average of squared gradients to adjust the learning rate automatically. This helps the model converge faster and makes RMSprop particularly effective for non-stationary and recurrent neural network problems.

**Adam (Adaptive Moment Estimation):**
Adam combines the advantages of both Momentum and RMSprop. It uses the moving averages of both gradients and squared gradients to adapt the learning rate for each parameter. Adam is computationally efficient, converges quickly, and works well on large datasets and deep neural networks. Due to its speed and reliability, it is one of the most widely used optimizers in deep learning.

**Difference between Gradient Descent, RMSprop, and Adam:**

* **Gradient Descent** uses a fixed learning rate for all parameters and is simple to implement, but it may converge slowly and is sensitive to the choice of learning rate.
* **RMSprop** adjusts the learning rate automatically for each parameter, resulting in faster convergence and better performance on complex problems.
* **Adam** combines adaptive learning rates with momentum, providing faster and more stable convergence than both Gradient Descent and RMSprop. It is generally the preferred optimizer for most deep learning applications.



**Question 6: Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.**

**Answer:**

```python
import numpy as np

# AND Gate Input and Output
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

# Initialize weights and bias
weights = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 10

# Step activation function
def activation(x):
    if x >= 0:
        return 1
    else:
        return 0

# Training the Perceptron
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = activation(linear_output)

        error = y[i] - prediction

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

# Testing the Perceptron
print("Predictions for AND Gate:")
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    prediction = activation(linear_output)
    print(f"{X[i]} -> {prediction}")
```

**Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.**

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-10, 10, 100)

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Plot the activation functions
plt.figure(figsize=(10,6))

plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, tanh(x), label='Tanh')

plt.title("Sigmoid, ReLU, and Tanh Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()

plt.show()
```

**Question 8: Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy.**

**Answer**


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize the pixel values
X_train = X_train / 255.0
X_test = X_test / 255.0

# Build the neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=32,
    verbose=1
)

# Print training accuracy
train_loss, train_accuracy = model.evaluate(X_train, y_train, verbose=0)

print("Training Accuracy:", train_accuracy)
```

**Question 9: Visualize the loss and accuracy curves for a neural network model trained
on the Fashion MNIST dataset. Interpret the training behavior.**

**Answer:**

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense

# Load the Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize the data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Build the neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    validation_data=(X_test, y_test),
    verbose=1
)

# Plot Loss Curve
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plot Accuracy Curve
plt.figure(figsize=(8,5))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

**Question 10: You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured
features like transaction amount, merchant ID, and customer location. The goal is to
classify each transaction as fraudulent or legitimate.
Explain your real-time data science workflow:**
* **How would you design a deep learning model (perceptron or multilayer NN)?**
* **Which activation function and loss function would you use, and why?**
* **How would you train and evaluate the model, considering class imbalance?**
* **Which optimizer would be suitable, and how would you prevent overfitting?**

**Answer:**

To build a real-time fraud detection system, the workflow should include data preprocessing, deep learning model development, model training, evaluation, and continuous monitoring. Since the dataset is large, imbalanced, and contains structured transaction data, a Multilayer Neural Network (MLP) is more suitable than a single-layer perceptron because it can learn complex relationships among the features.

**Step 1: Data Collection and Preprocessing**

The first step is to collect transaction data containing features such as transaction amount, merchant ID, customer location, and transaction history. The data is cleaned by handling missing values, encoding categorical variables, scaling numerical features, and splitting it into training, validation, and test datasets.

**Step 2: Deep Learning Model Design**

A Multilayer Neural Network (MLP) should be used because fraud detection is a complex classification problem. The network consists of an input layer, one or more hidden layers, and an output layer. Hidden layers enable the model to learn complex patterns and interactions between the transaction features, resulting in better classification performance than a single-layer perceptron.

**Step 3: Activation Function and Loss Function**

The ReLU (Rectified Linear Unit) activation function should be used in the hidden layers because it speeds up training, reduces the vanishing gradient problem, and performs well in deep neural networks. For the output layer, the Sigmoid activation function is appropriate because the task is binary classification (fraudulent or legitimate transaction). The Binary Cross-Entropy loss function should be used because it effectively measures the prediction error for binary classification problems.

**Step 4: Training and Evaluation**

Since fraudulent transactions are much fewer than legitimate ones, the dataset is imbalanced. To handle this imbalance, techniques such as class weighting, oversampling, or SMOTE (Synthetic Minority Over-sampling Technique) can be applied. The model should be evaluated using metrics such as Precision, Recall, F1-Score, ROC-AUC, and the Confusion Matrix, as accuracy alone may be misleading for imbalanced datasets.

**Step 5: Optimizer and Overfitting Prevention**

The Adam optimizer is the most suitable choice because it provides fast convergence, adapts the learning rate automatically, and performs well on large datasets. To prevent overfitting, techniques such as Dropout, Early Stopping, L2 Regularization, and using a validation dataset during training should be applied. These methods help the model generalize well to new transaction data.

**Step 6: Business Impact**

The deployed model can classify transactions as fraudulent or legitimate in real time, allowing suspicious transactions to be flagged immediately. This helps reduce financial losses, improve customer security, minimize manual review efforts, and support faster, data-driven decision-making for fraud prevention.
